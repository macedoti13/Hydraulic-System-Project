```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_parquet("../data/curated_datasets/water_consumption_curated.parquet")
df = df[["output_flow_rate", "total_precip_mm", "air_temp_c", "max_temp_last_hour_c", "min_temp_last_hour_c", "relative_humidity_percentage"]]
df = df.iloc[10000: 25000]
df.head()
```

```python
correlation_air_temp = df['output_flow_rate'].corr(df['air_temp_c'])
correlation_total_precip_mm = df['output_flow_rate'].corr(df['total_precip_mm'])
correlation_relative_humidity = df['output_flow_rate'].corr(df['relative_humidity_percentage'])

# Criao dos grficos de disperso usando plotly.graph_objects
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=df['air_temp_c'], y=df['output_flow_rate'], mode='markers', marker_color='purple'))
fig1.update_layout(title=f'Correlao entre Temperatura do Ar e Consumo de gua (r={correlation_air_temp:.2f})', xaxis_title='Temperatura do Ar (C)', yaxis_title='Consumo de gua (L/s)')

# Exibindo os grficos
fig1.show()
```

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=df['total_precip_mm'], y=df['output_flow_rate'], mode='markers', marker_color='red'))
fig1.update_layout(
    title=f'Correlao entre Precipitacao e Consumo de gua (r={correlation_air_temp:.2f})',
    xaxis_title='Precipitacao (mm)',
    yaxis_title='Consumo de gua (L/s)',
    height=600,
    width=800
)

# Exibindo os grficos
fig1.show()
```

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=df['relative_humidity_percentage'], y=df['output_flow_rate'], mode='markers', marker_color='green'))
fig1.update_layout(
    title=f'Correlao entre Humidade Relativa e Consumo de gua (r={correlation_air_temp:.2f})',
    xaxis_title='Humidade Relativa (%)',
    yaxis_title='Consumo de gua (L/s)',
    height=600,
    width=800
)

# Exibindo os grficos
fig1.show()
```
