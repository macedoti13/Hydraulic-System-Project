```python
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_parquet("../data/silver/water_consumption_silver.parquet")
df.head()
```

```python
df_weekdays = df[df['day_of_week'] < 5]
df_weekends = df[df['day_of_week'] >= 5]
```

```python
flow_out_weekdays = df_weekdays[df_weekdays.inputed_row==False].groupby('hour')['output_flow_rate_mean'].mean().reset_index().rename(columns={'output_flow_rate_mean': 'avg_weekday_output_flow'})
flow_out_weekends = df_weekends[df_weekends.inputed_row==False].groupby('hour')['output_flow_rate_mean'].mean().reset_index().rename(columns={'output_flow_rate_mean': 'avg_weekend_output_flow'})
```

```python
df_combined = pd.merge(flow_out_weekdays, flow_out_weekends, on='hour', how='outer')
```
