```python
from utils.scripts.functions.main_functions import forecast_next_24_hours_output_flow_rate
import pandas as pd
import pickle
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', None)
# Load the dataset and models
df = pd.read_parquet("data/curated_datasets/water_consumption_curated.parquet")
forecaster = pickle.load(open("models/forecaster.pkl", "rb"))
input_flow_model = pickle.load(open("models/input_flow_forecaster.pkl", "rb"))

# Forecast the next 24 hours
forecasted_df = forecast_next_24_hours_output_flow_rate(
    input_df=df, 
    forecaster=forecaster, 
    input_flow_model=input_flow_model,
    year=2023, 
    month=6, 
    day=15, 
    hour=12,
    include_weather_features=False
)
```
